## <center> **Feature <span style="color: red">Engineering</span>**

#### *Avg uploads per month*

```python
# We will take the diffrence between channel' s starting date and the data base creaton date.
base_channels["avg_uploads_per_month"] = base_channels["video_count"] / (
    
    (pd.Series(np.full(len(base_channels),
                      datetime.strptime("2032-4-16", '%Y-%m-%d'))).dt.to_period('M') -
    base_channels["date"].dt.to_period('M')).apply(lambda x: x.n)
).astype(np.float16)


base_channels_videos["avg_uploads_per_month"] = base_channels_videos["video_count"] / (
    
    (pd.Series(np.full(len(base_channels_videos),
                      datetime.strptime("2032-4-16", '%Y-%m-%d'))).dt.to_period('M') -
    base_channels_videos["date"].dt.to_period('M')).apply(lambda x: x.n)
).astype(np.float16)
```

```python
base_channels.head()
```

#### *Video title & describtion length*

```python
base_channels_videos["description_length"] = base_channels_videos["description"].str.len()
base_videos["description_length"] = base_videos["description"].str.len()
```

```python
base_channels_videos["title_length"] = base_channels_videos["title"].str.len()
base_videos["title_length"] = base_videos["title"].str.len()
```

```python
base_channels_videos["channel_name_length"] = base_channels_videos["channel_name"].str.len()
base_channels["channel_name_length"] = base_channels["channel_name"].str.len()
```

#### *Social media accounts*

```python
for df in [base_games, base_videos, base_channels_videos, base_channels]:
    
    if df is base_channels:
        col = "about"
    else:    
        col = "description"
        
    df["have_twitter_account"] = df[col].str.contains("twitter")
    df["have_instgram_account"] = df[col].str.contains("instagram")
    df["have_facebook_account"] = df[col].str.contains("facebook")
    df["have_twitch_account"] = df[col].str.contains("twitch")
```

```python
base_channels_videos.sample(5)
```

#### *Average video stats per channel* 

```python
stacked_channels_videos["video_count"][
    stacked_channels_videos["video_count"] == 0] += 1
# We will add one to avoid zero division error

stacked_channels_videos["avg_views_count"] = (
    stacked_channels_videos["total_views"] / stacked_channels_videos["video_count"]).astype(
        np.uint32)

stacked_channels_videos["avg_like_count"] = (
    stacked_channels_videos.groupby(
        "channelTitle")["likeCount"].transform("mean")).astype(np.uint32)
```

```python
unique_views = stacked_channels_videos.groupby(
    'channelTitle')['avg_views_count'].mean(
     ).reset_index(name='avg_views_count').rename(
     columns= {"channelTitle": "channel_name"})

unique_likes = stacked_channels_videos.groupby(
    'channelTitle')['avg_like_count'].mean(
     ).reset_index(name='avg_like_count').rename(
     columns= {"channelTitle": "channel_name"})

base_channels = pd.merge(base_channels, unique_likes, on='channel_name', how='left')
base_channels = pd.merge(base_channels, unique_views, on='channel_name', how='left')
```

#### *Analysis for categorical varibels*

```python
tags_dict = {"Tags": [], "Views": [], "Comments": [], "Likes": [],
             "Count": []}
total_views = []
all_tags_temp = base_videos["tags"].explode().explode().tolist()

for tags_chunk in all_tags_temp:
    
    for tag, views, comments, likes in zip(ast.literal_eval(tags_chunk), base_videos["viewCount"],
                                           base_videos["commentCount"], base_videos["likeCount"]):

        tags_dict["Tags"].append(tag)
        tags_dict["Views"].append(views)
        tags_dict["Comments"].append(comments)
        tags_dict["Likes"].append(likes)
        tags_dict["Count"].append(1)
    
tags_df = pd.DataFrame({"Tag": tags_dict["Tags"],
                        "Views": tags_dict["Views"],
                        "Comments count": tags_dict["Comments"],
                        "Likes": tags_dict["Likes"],
                        "Count": tags_dict["Count"]}).groupby('Tag').sum().astype({
    "Views": np.uint32,
    "Comments count": np.uint16,
    "Count": np.uint16,
    "Likes": np.uint32})

# Set the column tag again after we turned it into index
tags_df["Tag"] = tags_df.index
```

```python
top_20_tags = tags_df.sort_values(by= "Count", ascending= False).head(20)

fig = px.bar(top_20_tags, x= "Tag", y= "Count", 
             color_discrete_sequence= [THEME_COLORS[1]])

fig.update_layout(
    font_family= "Franklin Gothic",
    hovermode= "closest",
    title= "Tags appending count")

fig.add_annotation(
    text = ("We can find that the shorts in gaming now are more <br>" + \
            "popular now than lives."),
    showarrow= True,
    x = 15,
    y = 500,
    arrowhead=1,
    arrowsize=1,
    arrowwidth=2)

fig.add_annotation(
    text = ("Looks like people looks<br> for gameplay videos<br>" + \
            "more than funny ones."),
    x = 3,
    y = 950,
    showarrow= False)

fig.update_xaxes(
    linecolor='black',
    gridcolor='darkgrey')

fig.update_yaxes(
    linecolor='black',
    gridcolor='darkgrey')

save_fig(fig, "tags_appendings_per_vid_bar_chart")
iplot(fig)
```

#### *Geographic analysis*

```python
geographic_df = pd.merge(stacked_channels, base_channels, on= "channel_name",
                         suffixes= ["", "_"])

geographic_df.drop(["video_count_", "total_views_", "subscribers_"], inplace= True,
                   axis= 1)


outliers = tucky_method(geographic_df["subscribers"].to_numpy())
outliers_indexes = np.array(*np.where(np.isin(geographic_df["subscribers"], outliers)))
geographic_df = geographic_df.drop(outliers_indexes)

alpha2_codes = geographic_df['country'].unique().tolist()
alpha3_codes = []

for code in alpha2_codes:
    try:
        country = pycountry.countries.get(alpha_2=code)
        alpha3_codes.append(country.alpha_3)
    except AttributeError:
        alpha3_codes.append(None)

geographic_df['country'] = geographic_df['country'].cat.remove_unused_categories()
geographic_df['country'] = geographic_df['country'].apply(lambda x: alpha3_codes[alpha2_codes.index(x)])

geographic_df = geographic_df.drop(geographic_df[geographic_df.country == "N/A"].index).groupby(
    by=["country",
        "Collecting date"])[["subscribers", "video_count",
                             "total_views", "avg_like_count"]].mean().reset_index()

geographic_df['Collecting date'] = geographic_df['Collecting date'].apply(lambda x: str(x))

fig = px.choropleth(geographic_df, locations= "country",
                    animation_frame= "Collecting date",
                    color= "subscribers",
                    hover_name= "country",
                    color_continuous_scale= "Reds",
                    projection= "natural earth",
                    range_color=[geographic_df['subscribers'].min(),
                                 geographic_df['subscribers'].max()],
                    title="Subscribers mean per country")

fig.update_layout(
    font_family= "Franklin Gothic",
    hovermode= "closest",
    width=700,
    height=500,
    title= "<span style='color: red'>Subscribers</span> mean per country.")


save_fig(fig, "channels_subs_per_country")
iplot(fig)
```
