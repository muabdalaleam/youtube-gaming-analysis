```python
def solution(area: int) -> list:
    """This function takes area in squared units
    with integer class.
    @params: takes area in square units & should
    be int
    @returns: list of largest 4 squares"""
    
    areas: list = []
    
    # Checking if the area is integer
    if area % 1 == 0:
        for square in range(area, 0):
           side_length = square ** (1/2) 
           
           # Checking if the side length is int and areas not full
           if (side_length % 1 == 0) and (len(areas) != 4):
                areas.append(square)
                area -= square
                
        return areas
        
print(solution(15324))
```

<h3>4- Feature engineering

```python
continous_col = ["budget", "revenue", "popularity", "budget_adj",
                 "revenue_adj", "runtime", "vote_count", "vote_average"]
```

```python
# Dropping outliers using tucky method which
# is not sharp outlier detector
def tucky_method(df, col):
    """
    This function works with pandas data frame and it treats
    the outliers of the column
    :params:Takes the data frame and coumn name.
    :returns:the column without the outliers.
    """
    
    Q3 = np.quantile(df[col], 0.75)
    Q1 = np.quantile(df[col], 0.25)
    IQR = Q3 - Q1
    
    upper_range = Q3 + (IQR * 1.5)
    lower_range = Q1 - (IQR * 1.5)
    
    df.reset_index(drop = True)
    
    outlier_free_list = [x for x in df[col] if (
         (x > lower_range) & (x < upper_range))]
    filtered_data = df.loc[df[col].isin(outlier_free_list)]
            
    return filtered_data
```

```python
for col in continous_col:
    df.append(tucky_method(df, col), ignore_index= True)
df.head()
```

```python
df = df.drop(["id", "imdb_id", "homepage", "tagline", "keywords"], axis= 1)
```

```python
df.info()
```

```python
for col in ["cast", "director", "genres", "overview"]:
    
    df[col] = df[col].fillna(df[col].mode)
    df = df.join(df[df.columns.difference(df.columns)])  
    
# We aare forced to drop production companies nan's
df = df.dropna(axis= 0)
```

<h3>2- Finding the top movies

```python
def plot_top50_movies(col, color, title, ylabel):
    fig, ax = plt.subplots()
    top50_movies = df.sort_values(col,  ascending=[0]).head(50)

    ax.bar(x= top50_movies["original_title"],
           height= top50_movies[col],
           color= color)

    plt.xticks(rotation= 90)
    fig.set_size_inches(12, 6)
    plt.title(title)
    plt.ylabel(ylabel)
    plt.show()
```

```python
plot_top50_movies("revenue", theme_colors[0],
                  title= "Top films in revenue",
                  ylabel= "Scaled revenues")
# plt.savefig('best film in revenue.png')
```

```python
plot_top50_movies("budget", theme_colors[1],
                  title= "Top films in budget",
                  ylabel= "Scaled budgets")
# plt.savefig('best film in budget.png')
```

```python
plot_top50_movies("vote_average", theme_colors[2],
                  title= "Top films in vote rate",
                  ylabel= "Average votes from 10")
# plt.savefig('best films to thier rate.png')
```

<h5>From the last two steps we can conclude that:</h5><ul>
    <li>The highest revenue doesn't mean it should be the highest budget
    <li>The higest revenue or budget doesn't completly mean the higest rate
    <li>The vote count affect strongly the popularity of the film
    <li>Avatar was one of the higest revenues and budgets but not even in top 50 highest rate
    <li>Revenue is really affected by vote count even more than the budget

<h3>3- The highest directors in making revenues

```python
directors = df["director"].str.split("|")
directors_appearences = directors.value_counts(sort=True).head(50)
directors_appearences.index = directors_appearences.index.map(str)
```

```python
fig, ax = plt.subplots()

plt.bar(x= directors_appearences.index.values, height= directors_appearences,
        color= theme_colors[0])

plt.xticks(rotation= 90)
plt.ylabel("Count of films he shoot")
fig.set_size_inches(12, 5)
plt.title("Top directors in the count of thier films")
plt.show()
# plt.savefig('best directors.png')
```

<h3>4- Changes in the features by year

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols= 3)

df_by_year = df.groupby("release_year")
df_by_year["revenue"].mean().plot.area(y= "release_year", color= theme_colors[0],
                                       use_index= True, xlabel= "Release year",
                                       ylabel= "Scaled revenue", ax= ax1)

df_by_year["budget"].mean().plot.area(y= "budget", color= theme_colors[1],
                                       use_index= True, xlabel= "Release year",
                                       ylabel= "Scaled budget", ax= ax2)

df_by_year["vote_average"].median().plot.area(y= "vote_average", color= theme_colors[2],
                                       use_index= True, xlabel= "Release year",
                                       ylabel= "Vote average", ax= ax3)

ax1.set_title("Change in revenue mean by years")
ax2.set_title("Change in budget mean by years")
ax3.set_title("Change in films rate mode by years")

fig.set_size_inches(18, 5)
fig.suptitle("Changing in some features by time", fontsize=20)
plt.show()
```

### 5- Top watched films genres
